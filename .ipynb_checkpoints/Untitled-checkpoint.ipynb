```python
import pandas as pd
import os
from datetime import datetime, timedelta
from toad.plotter import plot_hist_prices
```

```python
start = datetime(2018, 1, 1)
end = datetime.today() - timedelta(1)

myPath = '/home/alessandro/Documenti/oanda_helper/data/market_data'

dates = pd.date_range(start, end)
instrument = 'XAG_CAD'
```

```python
mySerie = pd.DataFrame()
for date in pd.date_range(start, end):
    try:
        tmp = pd.read_pickle(os.path.join(myPath, instrument, f"{date.strftime('%Y%m%d')}_{instrument}.pickle"))
        mySerie = mySerie.append(tmp)
    except:
        pass
```

```python
plot_hist_prices(instrument, mySerie)
```
